```python
import numpy as np

# Create array
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)

# Vectorized operations (much faster than loops)
print("Squared:", arr ** 2)
print("Mean:", np.mean(arr))
print("Random 3x3 matrix:\n", np.random.rand(3, 3))

# Matrix multiplication
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print("Matrix product:\n", np.dot(a, b))
```

```python
import pandas as pd
import numpy as np

# Create DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
}
df = pd.DataFrame(data)
print("DataFrame:\n", df)

# Easy filtering and grouping
print("Salary > 55k:\n", df[df['Salary'] > 55000])
print("Average age:", df['Age'].mean())

# Group by (simulates SQL)
df['AgeGroup'] = pd.cut(df['Age'], bins=2)
print("\nGrouped by age:\n", df.groupby('AgeGroup', observed=False)['Salary'].mean(),)
```
