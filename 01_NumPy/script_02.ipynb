```python
# Random

rng = np.random.default_rng()  # Modern RNG (recommended)

print(rng.random(3))           # 1D array: [0.548, 0.715, 0.602] (varies)
print(rng.random((2, 3)))      # 2D: [[0.424, 0.658, 0.118], [0.908, 0.207, 0.232]]
print(rng.uniform(-1, 1, 5))   # Range -1 to 1: [-0.301, 0.956, -0.601, 0.418, 0.023]
```

```python
print(rng.integers(0, 10, 4))  # 0-9: [7, 4, 2, 9]
print(rng.integers(1, 7, (2,3)))  # Dice rolls: [[3, 6, 2], [4, 1, 5]]
```

```python
fruits = np.array(['apple', 'banana', 'cherry'])
print(rng.choice(fruits, 3))         # With replacement: ['banana', 'apple', 'cherry']
print(rng.choice(fruits, 2, replace=False))  # Without: ['cherry', 'apple']
```

```python
arr = np.array([1, 2, 3, 4, 5])
rng.shuffle(arr)         # In-place: [3, 1, 5, 2, 4]
print(rng.permutation([10, 20, 30]))  # New shuffled copy: [30, 10, 20]
```
