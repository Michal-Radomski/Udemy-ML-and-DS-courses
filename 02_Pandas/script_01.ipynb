```python
# Sample DataFrame from before
data = {'Name': ['Alice', 'Bob', 'Charlie'], 
        'Age': [25, 30, 35], 
        'City': ['NYC', 'LA', 'Chicago']}
df = pd.DataFrame(data)

# Single row by position (row 1 = Bob)
print(df.iloc[1])

# Multiple rows (rows 0-1)
print(df.iloc[0:2])

# Specific rows and columns (rows 0,2 and columns 0,1)
print(df.iloc[[0,2], [0,1]])
```

```python
# Set custom index
df_indexed = df.set_index('Name')
print(df_indexed.loc['Bob'])  # Row where Name == 'Bob'

# Range of labels
print(df_indexed.loc['Alice':'Bob'])

# Rows with conditions
print(df.loc[df['Age'] > 28])  # Filter by boolean mask
```

```python
# Multiple conditions
filtered = df[(df['Age'] > 25) & (df['City'] != 'NYC')]
print(filtered)

# Using query method
print(df.query('Age > 25 and City != "NYC"'))
```

```python
# Crosstabs
# Extended sample data with categories
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Department': ['HR', 'IT', 'HR', 'IT', 'Finance'],
    'Gender': ['F', 'M', 'F', 'F', 'M'],
    'Salary': [50000, 60000, 55000, 70000, 65000]
}
df = pd.DataFrame(data)

# Cross-tab Gender vs Department
ct = pd.crosstab(df['Gender'], df['Department'])
print(ct)
```

```python
ct_totals = pd.crosstab(df['Gender'], df['Department'], margins=True)
print(ct_totals)
```

```python
# Average salary by Gender and Department
ct_salary = pd.crosstab(df['Gender'], df['Department'], 
                       values=df['Salary'], aggfunc='mean')
print(ct_salary)
```

```python
# Cut method
# Sample data from crosstab example
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Department': ['HR', 'IT', 'HR', 'IT', 'Finance'],
    'Salary': [50000, 60000, 55000, 70000, 65000]
}
df = pd.DataFrame(data)

# Create salary bins
bins = [0, 55000, 65000, 100000]
labels = ['Low', 'Medium', 'High']
df['Salary_Bin'] = pd.cut(df['Salary'], bins=bins, labels=labels)
print(df[['Name', 'Salary', 'Salary_Bin']])
```

```python
# Merging Data
# Employees DataFrame
employees = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'DeptID': ['D001', 'D002', 'D001', 'D003'],
    'Salary': [50000, 60000, 55000, 70000]
})

# Departments DataFrame
departments = pd.DataFrame({
    'DeptID': ['D001', 'D002', 'D004'],
    'DeptName': ['HR', 'IT', 'Finance']
})

# Inner merge (default) - only matching DeptID
merged = pd.merge(employees, departments, on='DeptID')
print(merged)
```

```python
# Left join - all employees + matching departments
left_merged = pd.merge(employees, departments, on='DeptID', how='left')

# Right join - all departments + matching employees  
right_merged = pd.merge(employees, departments, on='DeptID', how='right')

# Outer join - all rows from both
outer_merged = pd.merge(employees, departments, on='DeptID', how='outer')
print(outer_merged)
```

```python
# Different key names
emps = pd.DataFrame({'EmpID': [1, 2], 'Name': ['Eve', 'Frank']})
depts = pd.DataFrame({'EmployeeID': [1, 3], 'Role': ['Manager', 'Analyst']})
multi_key = pd.merge(emps, depts, left_on='EmpID', right_on='EmployeeID')

# Column name conflicts (add suffixes)
merged_suffix = pd.merge(employees, employees, on='DeptID', 
                        suffixes=('_left', '_right'))
print(merged_suffix)
```
