```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 10)  # Better viewing
```

```python
data = {'Name': ['Alice', 'Bob', 'Charlie'], 
        'Age': [25, 30, 35], 
        'Salary': [50000, 60000, 70000]}
df = pd.DataFrame(data)
df  # Displays table automatically in Jupyter
```

```python
df.head()  # First 5 rows
df.info()  # Data types, missing values
df.describe()  # Stats summary
```
